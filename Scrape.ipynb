```python
import pandas as pd
import json

labels = ["Country", "New Cases Per Million", "New Deaths Per Million", "New Cases", "New Deaths"]

df = pd.read_csv("perMillion.csv", names=labels)
df = df.iloc[1:]

df["New Cases Per Million"] = df["New Cases Per Million"].astype(float)
df["New Deaths Per Million"] = df["New Deaths Per Million"].astype(float)
df["New Cases"] = df["New Cases"].astype(float)
df["New Deaths"] = df["New Deaths"].astype(float)
 
df[df["Country"] == "Austria"]
```
